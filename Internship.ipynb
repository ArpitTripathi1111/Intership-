```python
import pandas as pd
df = pd.read_csv(r"C:\Users\arpit\Downloads\car data.csv")
df

```

```python
#Remove duplicates
# By default, this drops rows where ALL columns have identical values to a previous row.
df_cleaned = df.drop_duplicates()

# 3. Export to a new Excel file
# index=False prevents pandas from writing row numbers into the new file
df_cleaned.to_excel('cleaned_data.xlsx', index=False)
df_cleaned
```

```python
df.info() # give information about data types
```

```python
df.replace({'Fuel_Type': {'Petrol': 1, 'Diesel': 2, 'CNG': 3}}, inplace=True)
df
```

```python
df
```

```python
df.replace({'Seller_Type': {'Dealer': 1, 'Individual': 2, 'Trustmark Dealer': 3, 'Arpit': 4}}, inplace=True) #Trustmark Dealer and Arpit are not present in any element type but still it is not giving any error
df
```

```python
df.replace({"Transmission": {'Manual': 1, 'Automatic': 2}}, inplace=True) 
df
```

```python
df.replace({"Year": {2014: 1, 2015: 2, 2016: 3}}, inplace=True) 
df
```

```python
x = df.drop(['Car_Name', 'Selling_Price'], axis=1)
y = df['Selling_Price']
```

```python
print(x)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# 1. Define your target variable Y
Y = df['Selling_Price']

# 2. Now perform the split
X_train, X_test, Y_train, Y_test = train_test_split(x, Y, test_size=0.1, random_state=2)
lin_reg_model = LinearRegression()
lin_reg_model.fit(X_train, Y_train)
```

```python
training_data_prediction = lin_reg_model.predict(X_train)
```

```python
from sklearn import metrics
error_score= metrics.r2_score(Y_train, training_data_prediction)
print("R squared error : ", error_score)
```

```python
import matplotlib.pyplot as plt
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual Prices")  
plt.ylabel("Predicted Prices")
plt.title("Actual Prices vs Predicted Prices")
plt.show()
```
